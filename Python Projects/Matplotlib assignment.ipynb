## Data Visualization of IPL (2008-2019)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_IPL2020=pd.read_excel('D:/BUSINESS ANLYTICS/PYTHON/Matches IPL 2008-2019.xlsx')
```

```python
df_IPL2020
```

```python
df_IPL2020.drop(['umpire3', 'dl_applied'],axis=1,inplace=True)
```

```python
df_IPL2020
```

```python
print (df_IPL2020['season'].unique())
```

```python
df_IPL2020.describe()
```

```python
matches_won_by_each_team_df = df_IPL2020['winner'].value_counts()
```

```python
matches_won_by_each_team_df
```

```python
matches_won_by_each_team_df.plot(kind="barh", title="Matches won (2008-2019)", color='y',figsize=(12,6))
plt.xlabel('no. of wins')
plt.ylabel("Teams")
plt.legend()
plt.show()
```

```python
df_IPL2020.iloc[df_IPL2020['win_by_runs'].idxmax()]
```

```python
df_IPL2020.iloc[df_IPL2020['win_by_wickets'].idxmax()]
```

```python
matches_played_df=df_IPL2020['season'].value_counts()
```

```python
matches_played_df
```

```python
matches_played_df.plot(kind='bar', figsize=(15,5), color= "g")
plt.title('Matches Played Each Season (2008-2019)')
plt.ylabel('Years')
plt.xlabel('no. of match')
plt.legend()
plt.show()
```

```python
matches_played_df.plot(kind='pie', figsize=(12,12),autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Matches Played Each Season (2008-2019)')
plt.legend(loc='upper left')
plt.show()
```

```python
stadium_venue_df=df_IPL2020['venue'].value_counts()
```

```python
stadium_venue_df
```

```python
stadium_venue_df.plot(kind='bar', figsize=(15,5), color='Brown')
plt.title('Number of Matches Played in Each Venue')
plt.ylabel('No. of matches')
plt.xlabel('Stadium venue')
plt.legend()
plt.show()
```

```python
city_venue_df=df_IPL2020['city'].value_counts()
```

```python
city_venue_df
```

```python
city_venue_df.plot(kind='bar',figsize=(15,5), color='orange')
plt.title('Number of matches held in a city')
plt.ylabel('No. of matches')
plt.xlabel('city name')
plt.legend()
plt.show()
```
