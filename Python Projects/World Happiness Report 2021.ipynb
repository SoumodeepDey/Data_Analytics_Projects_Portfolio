```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Ladder score of top & bottom 10 countries

# Top 10 countries
fig = plt.figure(figsize = (15,5))
x_1 = df['Country name'].head(10)
y_1 = df['Ladder score'].head(10)
plt.bar(x_1, y_1, color = 'Brown', label = 'Ladder Score', width = 0.45, edgecolor = 'k')
plt.title('Top 10 countries', fontsize = 25, fontname = 'Times New Roman')
plt.xlabel('Countries', fontsize = 15, fontname = 'Times New Roman')
plt.ylabel('Ladder score', fontsize = 15, fontname = 'Times New Roman')
plt.legend()
plt.show()

# Bottom 10 countries
fig = plt.figure(figsize = (15,5))
x_2 = df['Country name'].tail(10)
y_2 = df['Ladder score'].tail(10)
plt.bar(x_2, y_2, color = 'orange', label = 'Ladder Score', width = 0.45, edgecolor = 'k')
plt.title('Bottom 10 countries', fontsize = 25, fontname = 'Times New Roman')
plt.xlabel('Countries', fontsize = 15, fontname = 'Times New Roman')
plt.ylabel('Ladder score', fontsize = 15, fontname = 'Times New Roman')
plt.legend()
plt.show()
```

```python
# Regional wise standard error of ladder score


reg_std_err_ls = df.groupby(['Regional indicator'])[['Standard error of ladder score']].sum()

fig=plt.figure(figsize=(12,12))
size= reg_std_err_ls['Standard error of ladder score']
labels= df['Regional indicator'].unique()
colors=['#e68d60','#e6e660','#98e660','#60e688','#60b7e6','#9a60e6','#93afb5','#908791','#79b08f','#abb079']
my_circle = plt.Circle ((0,0),0.7,color='white')
plt.pie(size,colors=colors, labels=labels, shadow = True, autopct = '%.2f%%')
plt.title("Regional wise standard error of ladder score",fontsize=30, fontname = 'Times New Roman')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.legend(loc=4)
plt.show()
```

```python
# upper whisker & lower whisker of top / bottom 10 countries

fig = plt.figure(figsize= (15, 8))
con_df = df['Country name'].head(10)

uw_con = df['upperwhisker'].head(10)
lw_con = df['lowerwhisker'].head(10)

plt.barh(con_df, uw_con, 
         label= 'upperwhisker countries', 
         color =(['#e68d60','#e6e660','#98e660','#60e688','#60b7e6','#9a60e6','#93afb5','#908791','#79b08f','#abb079'])
        )
plt.title('Top 10 upper whisker Countries', fontsize = 25, fontname = 'Times New Roman')
plt.ylabel('Countries', fontsize = 15, fontname = 'Times New Roman')
plt.xlabel('upperwhisker', fontsize = 15, fontname = 'Times New Roman')
plt.legend()
plt.show()

fig = plt.figure(figsize= (15, 8))
plt.barh(con_df, lw_con,
         label= 'lowerwhisker countries',
         color=(['#e68d60','#e6e660','#98e660','#60e688','#60b7e6','#9a60e6','#93afb5','#908791','#79b08f','#abb079'])
        )
plt.title('Top 10 lower whisker Countries',fontsize = 25,fontname = 'Times New Roman')
plt.ylabel('Countries', fontsize = 15, fontname = 'Times New Roman')
plt.xlabel('lowerwhisker',  fontsize = 15,  fontname = 'Times New Roman')
plt.legend()
plt.show()
```

```python
# Logged GDP regional wise

plt.figure(figsize=(15,8))
sns.barplot('Logged GDP per capita', 'Regional indicator', data= df)
plt.title('Logged GDP regional wise', fontsize = 25, fontname = 'Times New Roman')

plt.xlabel('Logged GDP', fontsize = 15, fontname = 'Times New Roman')
plt.ylabel('Region', fontsize = 15, fontname = 'Times New Roman');
```

```python
# Social support on each region

a = df.groupby('Regional indicator')[['Social support']].sum()

labels = a.index
size = a['Social support']

colors=['#e68d60','#e6e660','#98e660','#60e688','#60b7e6','#9a60e6', '#98e123','#60e574','#62b7e0','#9a61e6']

plt.rcParams['figure.figsize']=(12,12)
plt.pie(size,colors=colors, labels=labels, shadow = True, autopct = '%.2f%%')

plt.title("Social support on each region",fontsize=30, fontname = 'Times New Roman')
plt.legend(loc= 'lower right');
```

```python
# How social support is affecting log GDP per region?

a = df.groupby('Regional indicator')[['Explained by: Log GDP per capita']].sum()
b = df.groupby('Regional indicator')[['Explained by: Social support']].sum()

plt.figure(figsize=(15,8))
plt.plot(df['Regional indicator'].unique(), a['Explained by: Log GDP per capita'].unique(), 's--r', mec='k', ms=8, mew=2, lw=3)
plt.plot(df['Regional indicator'].unique(), b['Explained by: Social support'].unique(), 'o--b', mec='k',ms=8, mew=2, lw=3)

plt.title('Social support vs log GDP per region', fontsize=30, fontname = 'Times New Roman')
plt.legend(labels=['Log GDP per region', 'Social support'])
plt.xlabel('Region', fontsize=15, fontname = 'Times New Roman')
plt.ylabel('Log GDP' ,fontsize=15, fontname = 'Times New Roman')
plt.xticks(rotation = 90);
```

```python
# How log GDP per region is affected by corruption?

a = df.groupby('Regional indicator')[['Explained by: Log GDP per capita']].sum()
b = df.groupby('Regional indicator')[['Explained by: Perceptions of corruption']].sum()

plt.figure(figsize=(15,8))

plt.bar(df['Regional indicator'].unique(), a['Explained by: Log GDP per capita'].unique(), color = 'orange',  width = 0.45, edgecolor = 'k')
plt.plot(df['Regional indicator'].unique(), b['Explained by: Perceptions of corruption'].unique(), 'o--g', mec='k',ms=8, mew=2, lw=3)

plt.title('log GDP vs corruption ', fontsize=30, fontname = 'Times New Roman')
plt.xlabel('Region', fontsize=15, fontname = 'Times New Roman')
plt.ylabel('Log GDP' ,fontsize=15, fontname = 'Times New Roman')
plt.legend(labels=['region wise Corruption', 'log GDP per region'])
plt.xticks(rotation = 90);
```

```python
# Which region has the best healthy life expectancy?

a = df.groupby('Regional indicator')[['Healthy life expectancy']].sum()

plt.figure(figsize=(15,8))

sns.barplot(a.index, a['Healthy life expectancy'].unique())

plt.title('Healthy life expectancy region wise', fontsize=30, fontname = 'Times New Roman')
plt.xlabel('Region', fontsize=15, fontname = 'Times New Roman')
plt.ylabel('Healthy life expectancy' ,fontsize=15, fontname = 'Times New Roman')
plt.xticks(rotation = 90);
```

```python
#  Which region has the highest freedom to make life choices?

a = df.groupby('Regional indicator')[['Freedom to make life choices']].sum()

plt.figure(figsize=(15,8))

sns.barplot(a.index, a['Freedom to make life choices'].unique(), alpha=0.5)
plt.title('Freedom to make life choices region wise', fontsize=30, fontname = 'Times New Roman')
plt.xlabel('Region', fontsize=15, fontname = 'Times New Roman')
plt.ylabel('Freedom rate' ,fontsize=15, fontname = 'Times New Roman')
plt.xticks(rotation = 90);
```
