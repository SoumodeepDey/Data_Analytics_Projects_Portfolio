### 12) Department wise attrition of the employee's

```python
bva_1 = pd.crosstab(IBM_df['Department'], IBM_df['Attrition'])
a=bva_1.div(bva_1.sum(1).astype(float), axis = 0).plot(kind = 'bar', stacked = False, figsize = (20, 10), color = ['#14dbb7', '#db146e','#8af298'])
plt.title('Department wise attrition',fontsize=30,fontname = 'Times New Roman')
plt.xlabel('Deparments', fontsize=20)
plt.xticks(rotation = 40, fontsize=15)
plt.legend(fontsize=15)
for p in a.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    a.annotate(f'{height:.0%}', (x + width/2, y + height*1.02), ha='center')
plt.show()
```

##### Observation :
- In sales deparment many of the employees have resigned from the sales (i.e. 21%)
- Many employees have also resigned from Human Resoures and Research & Development (i.e. 19% & 14%)
