```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Importing the data

df = pd.read_csv('D:/BUSINESS ANLYTICS/PYTHON/TravelInsurancePrediction.csv')
df
```

```python
# Dropping 1 unwanted column 

df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
# Shape of the data

df.shape
```

```python
# Describtion about the data

df.describe().transpose()
```

```python
# Checking any null values

df.isnull().any()
```

```python
# From which age group most of the application have been submited?

fig = plt.figure(figsize=(15,8))

df['Age'].plot.hist(bins=[20,25,30,35,40,45], color='#8aeaf2')

plt.xlabel('Age', fontsize=15, fontname = 'Times New Roman')
plt.ylabel("Frequency",fontsize=15, fontname = 'Times New Roman')
plt.title("Most age group applications",fontsize=30,fontname = 'Times New Roman');
```

```python
# Which type of employeed people are travelling the most?

fig = plt.figure(figsize=(15,8))

sns.countplot(df['Employment Type'], palette='Set2')

plt.xlabel('Employment type', fontsize=15, fontname = 'Times New Roman')
plt.ylabel("no. of people",fontsize=15, fontname = 'Times New Roman')
plt.title("Most travelling people",fontsize=30,fontname = 'Times New Roman');
```

```python
# What is the annual income of the people those who use to travel?

fig = plt.figure(figsize=(15,8))

df['AnnualIncome'].plot.hist(color='#d98032')

plt.xlabel('Annual income (annual income/100000)', fontsize=15, fontname = 'Times New Roman')
plt.ylabel("Frequency",fontsize=15, fontname = 'Times New Roman')
plt.title("Average annual income",fontsize=30,fontname = 'Times New Roman');
```

```python
# How many family member they preferred for travelling?

fig = plt.figure(figsize=(15,8))

sns.countplot(df['FamilyMembers'])

plt.xlabel('Age', fontsize=15, fontname = 'Times New Roman')
plt.ylabel("Count",fontsize=15, fontname = 'Times New Roman')
plt.title("Most no. of family members preferred",fontsize=30,fontname = 'Times New Roman');
```

```python
# Making few changes in the data

df['ChronicDiseases'] = df['ChronicDiseases'].map({1:'Yes', 0:'No'})
df['TravelInsurance'] = df['TravelInsurance'].map({1:'Yes', 0:'No'})
```

```python
# Are the travelers suffering from chronic diseases?

fig = plt.figure(figsize=(15,8))

sns.countplot(df['ChronicDiseases'], palette='Blues')

plt.xlabel('Chronic Diseases', fontsize=15, fontname = 'Times New Roman')
plt.ylabel("Count",fontsize=15, fontname = 'Times New Roman')
plt.title("Suffering from chronic diseases",fontsize=30,fontname = 'Times New Roman');
```

```python
# Are the applicants are frequent flyers?

fig = plt.figure(figsize=(15,8))

sns.countplot(df['FrequentFlyer'], palette='Greens')

plt.xlabel('Frequent flyer', fontsize=15, fontname = 'Times New Roman')
plt.ylabel("Count",fontsize=15, fontname = 'Times New Roman')
plt.title("Frequently Fly",fontsize=30,fontname = 'Times New Roman');
```

```python
# Are the applicants have ever travelled abroad? 

fig=plt.figure(figsize=(10,10))

size= df['EverTravelledAbroad'].value_counts()
labels= df['EverTravelledAbroad'].unique()
colors=['#60b7e6','#9a60e6']

my_circle = plt.Circle ((0,0),0.7,color='white')
plt.pie(size, colors=colors, labels=labels, shadow = True, autopct = '%.2f%%')

plt.title("Travelled Abroad",fontsize=30, fontname = 'Times New Roman')

p=plt.gcf()
p.gca().add_artist(my_circle)
plt.legend()
plt.show()
```

```python
# The applicants are having there travelling insurence or not?

fig = plt.figure(figsize=(15,8))

sns.countplot(df['TravelInsurance'], palette='hot')

plt.xlabel('Travel Insurence', fontsize=15, fontname = 'Times New Roman')
plt.ylabel("Count",fontsize=15, fontname = 'Times New Roman')
plt.title("Travelling Insurence Preference ",fontsize=30,fontname = 'Times New Roman');
```
